```python
import itertools

def simplify(swap_list):
    swap_list = sorted(swap_list)
    simple_list = []
    for k, g in itertools.groupby(swap_list):
        if len(list(g)) % 2:
            simple_list.append(k)
    return simple_list
        
```

```python
def swap_elems(l, ind1, ind2):
    p1, p2 = (l[ind1], l[ind2])
    l[ind1] = p2
    l[ind2] = p1
    return l
```

```python
def dance(programs, moves):
    programs = programs[:]
    for move in moves:
        if move[0] == 's':
            n = int(move[1:])
            programs = programs[-n:] + programs[:-n]
        elif move[0] == 'x':
            index1, index2 = [int(i) for i in move[1:].split('/')]
            programs = swap_elems(programs, index1, index2)
        elif move [0] == 'p':
            index1, index2 = (programs.index(move[1]), programs.index(move[3]))
            programs = swap_elems(programs, index1, index2)
    return programs
```

```python
def calculate_cycles(programs, moves):
    perms = {}
    initial_config = programs[:]
    permutation = dance(initial_config, moves)
    for i in range(len(initial_config)):
        index = initial_config.index(permutation[i])
        perms[i] = index
    return perms
```

```python
def calculate_swaps(programs, moves):
    swaps = {}
    initial_config = programs[:]
    permutation = dance(initial_config, moves)
    for i, p in enumerate(initial_config):
        swaps[p] = permutation[i]
    return swaps
```

```python
def do_partners(programs, moves):
    programs = programs[:]
    for move in moves:
        index1, index2 = (programs.index(move[1]), programs.index(move[3]))
        programs = swap_elems(programs, index1, index2)
    return programs
```

```python
def run(programs, moves, iterations):
    non_partners = [move for move in moves if move[0] != 'p']
    partners = [move for move in moves if move[0] == 'p']
    perms = calculate_cycles(programs, non_partners)
    after_perm = dance(programs, non_partners)
    swaps = calculate_swaps(after_perm, partners)
    perm_results = [perms[i] for i in range(len(programs))]
    for i in range(iterations):
        programs = [programs[res] for res in perm_results]
        programs = [swaps[p] for p in programs]
        if i % 1000000 == 0:
            print(i)
    return programs
```
